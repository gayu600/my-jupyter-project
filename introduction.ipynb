```python
# 11. 50% Discount is is offered on 200 Rs Pizza by "ABC firm on which 5% GST us applicable"
# if ordered through delievery partner.
# create a final bill for pizzas ordered by User.

#Taking Input from User
q = int(input("Enter Quantity :"))
#Expressions
total =  200 * q
disc = 50 / 100 * total
gst = 5/100 *(total - disc)
final = total - disc + gst

# Display Output
print("\n")
print("---------------- ~ Final Bil ~ ------------------")
print("\t\t Quantity Ordered    :",q)
print("\t\t Total Bill          :",total)
print("\t\t 50% Disc            :",disc)
print("\t\t 5% GST              :",gst)
print("\t\t Final Bill          :",final)
```

```python
# 14. WAP to display returns on SIP.
# Taking Input from User.
P = int(input("Enter Monthly Investment :"))
R = float(input("Enter Yearly RDI :"))
N = float(input("Enter Time Period In Years :"))
# Expressions
# Convert Yearly ROI into Monthly ROI
r = R / 12 / 100
#Convert Years into Months
n = N * 12
A = P * (((1 + r) ** n) - 1) * (1+r) / r
#Display Outcomes
print("\n")
print("\n---------------- ~ Final ~ ------------------")
print("Invested Amount  :", P * n )
print("Estimated Return :", A - (P * n))
print("Final Amount     : ",round(A,2))
```

```python
# 15. WAP to Implement EMI calculator.
# Taking Input from User.
P = int(input("Enter Monthly Investment :"))
R = float(input("Enter Yearly RDI :"))
N = float(input("Enter Time Period In Years :"))
# Expressions
# Convert Yearly ROI into Monthly ROI
r = R / 12 / 100
#Convert Years into Months
n = N * 12
E = P * r * ((1 + r) ** n) / ((1 + r) ** n) - 1
# Display Outcomes
print("\n")
print("\n---------------- ~ Final ~ ------------------")
print("Monthly EMI :", round(E,2) )
print("Total Amount :", E * n)
print("Interest     : ",(E * n)- P)
```
