```python
# 'copy' : Return a snallow copy of a set

a1 = { 101, 102, 103, 104, 105}

b1 = a1.copy()
print(b1)
```

```python
# Clear()

b1.clear()
print(b1)
```

```python
# Disjoint()

a1 = {34,34,67}
a2 = {56,89,20}
a1.isdisjoint(a2)
```

```python
# issuperset(), issubset()

a1 = {34,34,67,77,56,20,89}
a2 = {56,89,20}
print(a1.issuperset(a2))
print(a2.issubset(a1))
```

```python
# Union()

a1 = {34,34,67,77,56,20,89}
a2 = {56,89,20}

a3 = a1.union(a2)
print(a3)
```

```python
# Intersection()

a1 = {34,34,67,77,56,20,89}
a2 = {56,89,20}
a1.intersection(a2)
```

```python
# Difference()

a1 = {34,34,67,77,56,20,89}
a2 = {56,89,20}

print(a1.difference(a2))
```

```python
# Symmetric Difference()

a1 = {34,34,67,77,56,20,89}
a2 = {56,89,20}

print(a1.symmetric_difference(a2))
```
