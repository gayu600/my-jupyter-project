```python
# List is mutable

# Addition : 'append','insert','extend'
# Update  :
# Delete  : 'pop', 'remove'
```

```python
# Copy()  : Return a shallow copy of the list.
a1 = [11,22,33,44,55]
a2 = a1.copy()
print(a2)
```

```python
# clear()    : remove the element from the list

a1 = [11,22,33,44,55]

a1.clear()
print(a1)
```

```python
# index()  : Return first index of value.

a1 = [11,22,33,44,55]

a1.index(11)
```

```python
# Count :
a1 = [11,22,33,44,55,66,33,55,11,11]
a1.count(11)
```

# addition :

```python
# append() : Append object to the end of the list.
a1 = [11,22,33,44,55]
a1.append(55)
print(a1)
```

```python
# extend()

a1 = [11,22,33,44,55]
a1.extend([11,222,"fgh",555])
print(a1)
```

```python
# insert ()
a1 = [11,22,33,55,33]
a1.insert(2,"asd")
print(a1)
```

# deletion functions :

```python
# pop
a1 = [11,22,33,44,55]
a1.pop(-1)
print(a1)
```

```python
# remove
a1 = [11,22,33,44,55]
a1.remove(33)
print(a1)
```

```python
# Reverse   : mutable change the all data
a1 = [11,22,33,44,55]

a1.reverse()
print(a1)
```

```python
# Sort  : 
a1 = [11,22,333,441,567,3221,44,55]
a1.sort()
print(a1)
```

```python
# sort by desecnding
a1 = [12,353,543,65,6767,86767,677]
a1.sort(reverse = True)
print(a1)
```

## List Comprehension:

```python
# It is an elegant way of creating new lists.
```

```python
# Syntax : list_name =[               for i in         seq]
```

```python
# Create a List of Squares of numbers from 1 to 100.

b1=[i **2     for i in range(1,100)]
print(b1)
```

```python
a1 = []
for i in range (1,101):
    a1.append(i ** 2)

print(a1)
```

```python
# ii. table of given numbers.

n = int (input("Enter a nummber  :"))
tb = [n*i   for i in range(1,11)]
print(tb)
```

```python
# Create separate list of 
# i. Pass marks
# ii . Fail Students
# iii. marks in the range of 55 to 77.
# iv. result Pass/ Fail.
```

```python
# i. Pass marks
m1 = [23, 34, 35,38,45, 56, 67, 78, 89,39, 98, 87, 76, 65, 54, 43, 32, 21, 22, 33, 44, 77, 88, 99]
p = [ i       for i in m1   if i >=40]
print(p)
```

```python
# ii . Fail Students
m1 = [23, 34, 35,38,45, 56, 67, 78, 89,39, 98, 87, 76, 65, 54, 43, 32, 21, 22, 33, 44, 77, 88, 99]
p = [i    for i in m1   if i<=40]
print(p)
```

```python
# iii. marks in the range of 55 to 77.
m1 = [23, 34, 35,38,45, 56, 67, 78, 89,39, 98, 87, 76, 65, 54, 43, 32, 21, 22, 33, 44, 77, 88, 99]
n = [i   for i in m1  if i >=55 and i <= 77]
print(n)
```

```python
# iv. result Pass/ Fail.
m1 = [23, 34, 35,38,45, 56, 67, 78, 89,39, 98, 87, 76, 65, 54, 43, 32, 21, 22, 33, 44, 77, 88, 99]

R =["pass"  if i>=40 else "fail"     for i in m1 ]
print(R)
```

```python
# >66 : dist    >=60 % : first    >40%  : Second Class

m1 = [23, 34, 35,38,45, 56, 67, 78, 89,39, 98, 87, 76, 65, 54, 43, 32, 21, 22, 33, 44, 77, 88, 99]

R1 = [ "Distinction" if i>=60 else  "first class" if i<=60  else "Second Class"for i in m1]

# for loop using :
for i, j  in zip(m1,R1):
    print(i, "\t",j)
print(R1)
```
