```python
# Lambda functions is often used with map, filter, reduce.

# map : to write expressions for all elements.
# filter : conditions
# reduce : aggregations
```

```python
# 1. map()
# map(func, *iterables)
```

```python
# 1. WAP Lambda function to square all numbers.
a = [1,2,3,4,5,67,8,9,10]
s= map(lambda i: i **2,a)
print(list(s))
```

```python
# 2. WAP Lambda function to square all numbers.
Marks = [4,3,45,9,35,4,5,4,65,7,68,7,98,6,6,4,4,23,2,44,6,3,88,9]
marks = map(lambda i : i+ 1, Marks)
print(list(marks))
```

```python
# filter (function, iterable)
```

```python
# filter()
```

```python
# WA lambda function to filter pass:
Marks = [4,3,45,9,35,4,5,4,65,7,68,7,98,6,6,4,4,23,2,44,6,3,88,9]
p = filter(lambda i : i>= 40, Marks)
print(list(p))
```

```python
# Fail Marks
Marks = [4,3,45,9,35,4,5,4,65,7,68,7,98,6,6,4,4,23,2,44,6,3,88,9]
f = filter(lambda i : i< 40, Marks)
print(list(f))
```

```python
# WA filter function from square all even numbers.
p1 = [4,3,45,9,35,4,5,4,65,7,68,7,98,6,6,4,4,23,2,44,6,3,88,9]

e = map(lambda x : x **2 ,  filter(lambda i : i%2 ==0,p1))
print(list(e))
```

```python
import functools as ft
```

```python
# ft.reduce(function, iterable)
```

```python
# add all numbers.
a1 = [4,3,45,9,35,4,5,4,65,7,68,7,98,6,6,4,4,23,2,44,6,3,88,9,99,78]

s = ft.reduce(lambda s,i : s+i, a1)
s
```
