```python
# Genarate Marks of 50 student for 5 subjects.
np.random.randint(20,100,250).reshape([50,5])
```

```python
# Display first 5 rows.
# Display last 5 rows.
# Display rows from 11 to 20.
# Display subject wise toppers.
# Display  lowest marks by first 10 students.
# Display total makrs of all students.
# Display their % Marks.
# Identify how many are pass in each subject.
# Identify how many fail in 3 subject.
# Identify average marks in each subject.
# average marks in each subject of only pass students.
```

```python
marks = np.random.randint(20,100,250).reshape([50,5])
marks
```

```python
# Identify how many are pass in each subject.
marks[((marks [:, 0]>=40) &(marks[ :, 1]>=40) &(marks[ : ,2]>=40) & (marks[ :, 3]>=40)  &(marks[ :, 4]>=40))]
```

```python
print("count fo pass in subject 1 :",(marks[ :,0] >= 40).sum())
print("count fo pass in subject 2 :",(marks[ :,1] >= 40).sum())
print("count fo pass in subject 3 :",(marks[ :,2] >= 40).sum())
print("count fo pass in subject 4 :",(marks[ :,3] >= 40).sum())
print("count fo pass in subject 5 :",(marks[ :,4] >= 40).sum())
```

```python
# Identify average marks in each subject.
marks.mean(0)
```

```python
print("count fo pass in subject 1 :",(marks[ :,0] >= 40).sum())
print("count fo pass in subject 2 :",(marks[ :,1] >= 40).sum())
print("count fo pass in subject 3 :",(marks[ :,2] >= 40).sum())
print("count fo pass in subject 4 :",(marks[ :,3] >= 40).sum())
print("count fo pass in subject 5 :",(marks[ :,4] >= 40).sum())
```

```python
marks[marks[ :,0]>= 40][ :,0].mean().round()
```

```python
marks[marks[ :,1]>= 40][ :,1].mean().round(1)
```

```python
print("count fo pass in subject 1 :",(marks[ :,0] >= 40).mean().round())
print("count fo pass in subject 2 :",(marks[ :,1] >= 40).mean().round())
print("count fo pass in subject 3 :",(marks[ :,2] >= 40).mean().round())
print("count fo pass in subject 4 :",(marks[ :,3] >= 40).mean().round())
print("count fo pass in subject 5 :",(marks[ :,4] >= 40).mean().round())
```
