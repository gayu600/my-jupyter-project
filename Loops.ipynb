```python
# 1. WAP to display sum of all digits.

n= int(input("Enter a Number :"))
s = 0 
while n>0:
    d = n % 10 
    s = s + d
    n = n // 10
print("Sum Of Digits :",s)
```

```python
# 2. WAP to display Squares of sum of all digits.
n= int(input("Enter a Number :"))
s = 0 
while n>0:
    d = n % 10 
    s = s + (d ** 2)
    n = n // 10
print("Sum Of Digits :",s)
```

```python
# 3. WAP to display cubes of sum of all digits.

n= int(input("Enter a Number :"))
s = 0 
while n>0:
    d = n % 10 
    s = s + (d ** 3)
    n = n // 10
print("Sum Of Digits :",s)
```

```python
# 4. WAP to reverse a Number.

n = int(input("Enter a  Number :"))
rev = 0
while n > 0 :
    d = n% 10 
    n = n // 10 
 # reverse Number
    rev = rev * 10 + d
print(rev)
```

```python
# 5. WAP to check whether given number is pallindrome. 
n = int(input("Enter a  Number :"))
print("Original Number :",n)
n1 = n
rev = 0
while n > 0 :
    d = n% 10 
    n = n // 10 
 # reverse Number
    rev = rev * 10 + d
print("Reverse Number :",rev)
# to check whether pallindrome
if n1== rev:
    print("The number is pallindrom")
else :
    print("the number is not pallindrom")
```

```python
# 6. WAP to check whether given number is amstrong number
n1 = int(input("Enter a Number :"))
n2 = n1
sum = 0

while n1> 0:
    d = n1 % 10
    n1 = n1 // 10
    sum = sum + (d** 3)
print(sum)
if sum == n2:
    print("amstrong number")
else:
    print("not an amstrong Number")
```
