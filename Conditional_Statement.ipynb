```python
# WAP to implement FD calculator in HDFC bank.

# Taking output from users.
P = int(input("Enter Principal :"))
N = int(input("Enter Tenure in days :"))

# Condition
if N >= 7 and N <= 29 :
    R = 2.75
elif  N <= 45 :
    R = 3.25
elif  N <= 180 :
    R = 4.25
elif   N <= 270 :
    R = 5.5
elif  N <= (1 * 365) :
    R = 5.75
elif  N <= (15 * 30) :
    R = 6.25
elif  N <= (18 * 30) :
    R = 6.35
elif  N <= (3 * 365) :
    R = 6.45
elif  N <= (5 * 365) :
    R = 6.40
elif  N <= (10 * 365) :
    R = 6.15
# Expression
CI = P * (1 + R / 100) ** (N / 365)
  
# Display Output
print("\n\n")
print(" Principal Amount : ", P)
print(" Tenure           : ", N, "Days")
print(" Rate of Interest : ", R, "%")
print(" Interest Earned  : ", CI-P)
print(" Final Amount     : ", CI)
```

```python
# WAP to calculate Income Tax as per given slabs

# Up to ₹3,00,000: Nil
# ₹3,00,001 to ₹6,00,000: 5%
# ₹6,00,001 to ₹9,00,000: 10%
# ₹9,00,001 to ₹12,00,000: 15%
# ₹12,00,001 to ₹15,00,000: 20%
# Above ₹15,00,000: 30

# Taking input from user
inc = int(input("Enter Income from all Sources: "))

# Condition
if inc <= 300000:
    tax = 0

elif inc <= 600000:
    tax = 5/100 * (inc - 300000)

elif inc <= 900000:
    tax = 15000 + 10/100 * (inc - 600000)

elif inc <= 1200000:
    tax = 45000 + 15/100 * (inc - 900000)

elif inc <= 1500000:
    tax = 90000 + 20/100 * (inc - 1200000)

else:   # Above 15,00,000
    tax = 150000 + 30/100 * (inc - 1500000)

print("Income Tax Payable = ₹", tax)
```

```python
# WAP to display Salary Slip.

#      bs      da     hra    ta    fixed
#     <5000   70%     50%    60%    5000
#     <10000  60%     40%    50%    10000
#     <20000  60%     35%    65%    20000
#      else   70%     50%    50%    30000

# Taking input from the user
bs = int(input("Enter Basic Salary :"))

# condition
if bs < 5000:
    da = 0.70 * bs
    hra = 0.50 * bs
    ta = 0.60 * bs
    fixed = 5000
elif bs > 10000:
    da = 0.60 * bs
    hra = 0.40 * bs
    ta = 0.50 * bs
    fixed = 10000
elif bs < 20000:
    da = 0.60 * bs
    hra = 0.35 * bs
    ta = 0.65 * bs
    fixed = 20000
else:
    da = 0.70 * bs
    hra = 0.50 * bs
    ta = 0.50 * bs
    fixed = 30000
# Expression
gross = bs + da + hra + ta + fixed

# Display output
print("\n----- Salary Slip -----")
print("Basic Salary :", bs)
print("DA           :", da)
print("HRA          :", hra)
print("TA           :", ta)
print("Fixed Allow. :", fixed)
print("------------------------")
print("Gross Salary :", gross)
```
